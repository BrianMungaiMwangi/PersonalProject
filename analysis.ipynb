```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('travel insurance.csv')
df1.head()
```

```python
df2 = pd.read_csv('insurance.csv')
df2.head()
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2 = pd.DataFrame(df2)

df2.rename(columns={'age': 'Age'}, inplace=True)
print("\nDataFrame after renaming the column:")
print(df2)

```

```python
merged_df = pd.merge(df1, df2, on='Age', how='inner')

print(merged_df)
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

```python
merged_df.isna().sum()
# the number of NA record is significantly large compare to the total records. 
# Therefore, dropping all NA records is not recommended. Checking of NA values
```

```python
merged_df.loc[merged_df['Gender'].isnull()]
# There are 1152161 records don't have gender information, it could be struggling to do EDA with Gender attribute
# Drop all na values
```

```python
merged_df['Gender'].fillna('Unknown', inplace=True)
merged_df['Gender'].unique()
```

```python
#Checking duplicate values
print('There are', merged_df.duplicated().sum(),'duplicated records')
```

```python
merged_df = merged_df.loc[~merged_df.duplicated()]
```

```python
merged_df.describe()
```

```python
merged_df.info()
```

```python

bins = list(range(0,80,5))
bins
labels = []
for i in range(len(bins)-1):
    label = '{} - {}'.format(bins[i],bins[i+1])
    labels.append(label)
merged_df['Age_Group'] = pd.cut(merged_df['Age'], bins=bins, labels=labels, right=False)
merged_df
```

```python
#Checking distribution of Quantitative Attributes
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
merged_df.hist(ax=ax)
plt.tight_layout()
```

```python
def bar_plot(dataset):
    fig, ax = plt.subplots(figsize=(7,5))
    sns.barplot(data = dataset, y=dataset.index, x=dataset.columns[0])
    ax.bar_label(ax.containers[0])

top_agency = merged_df.groupby(merged_df['Agency']).agg({'Commision (in value)':'sum'}).nlargest(5, 'Commision (in value)')

bar_plot(top_agency)
```

```python
#Highest Travel Insurance Plan Sold table
plan_sold = merged_df.groupby(['Agency','Product Name']).size().reset_index(name="Count")
highest_plan_sold=plan_sold.nlargest(10,['Count'])
df_plan_sold_sort = highest_plan_sold.sort_values(by='Count')
df_plan_sold_sort
```

```python
#Top 5 Highest Travel Insurance Plan
best_sold_product = merged_df.groupby(merged_df['Product Name']).agg({'Product Name':'count'}).nlargest(5, 'Product Name')

bar_plot(best_sold_product)
```

```python
# Top 5 highest profit from the various products being sold
highest_profit_product = merged_df.groupby(merged_df['Product Name']).agg({'Commision (in value)':'sum'}).nlargest(5, 'Commision (in value)')

bar_plot(highest_profit_product)
```

```python
#Top 5 destination count
most_visited_country = merged_df.groupby(merged_df['Destination']).agg({'Destination':'count'}).nlargest(5, 'Destination')

bar_plot(most_visited_country)
```

```python
most_visited_region = merged_df.groupby(merged_df['region']).agg({'region':'count'}).nlargest(5, 'region')

bar_plot(most_visited_region)
```

```python
merged_df[["Age"]].describe()
```

```python
#Distribution of age
age_group = merged_df.groupby(merged_df['Age_Group']).agg({'Age_Group':'count'}).nlargest(10, 'Age_Group')

bar_plot(age_group)
```

```python
#Top 5 ages where insured are raising claim requests
df_claim=merged_df.loc[merged_df['Claim']=='Yes']
df_age= df_claim.groupby(by=["Age"]).size().reset_index(name="counts")

df_n_largest=df_age.nlargest(5,['counts'])

plt.pie(df_n_largest['counts'], labels=df_n_largest['Age'], autopct='%1.1f%%')
plt.title('Top 5 ages where insured are raising claim requests')
plt.show()
```

```python
#Top 5 countries where insured are raising claim requests
yes_claim = merged_df.loc[merged_df['Claim']=='Yes']
destination_claim= yes_claim.groupby(by=["Destination"]).size().reset_index(name="counts")
df_n_largest=destination_claim.nlargest(5,['counts'])

plt.pie(df_n_largest['counts'], labels=df_n_largest['Destination'], autopct='%1.1f%%')
plt.title('Top 5 countries where insured are raising claim requests')
plt.show()
```

```python
#A significant population of the travel insurance customers is unknown. 
#In contrast, the numbers for both the male and female are more less equal.
sns.countplot(merged_df,x='Gender')
plt.tight_layout()
```

```python
dist_channel = merged_df.groupby(by=['Distribution Channel']).size().reset_index(name="Count")
highest_dist=dist_channel.nlargest(10,['Count'])
df_dist_sort = highest_dist.sort_values(by='Count')
df_dist_sort
```

```python
#The substantial number of travel insurance agencies have online distribution channels, 
#with very few having a distribution channel in offline mode.
sns.countplot(merged_df,x='Distribution Channel')
plt.tight_layout()
```

```python
#Top 5 Agencies with Highest Online Distribution Channel
dist_agency = merged_df.groupby(['Agency', 'Distribution Channel']).size().reset_index(name='Count')
online_agencies = dist_agency[dist_agency['Distribution Channel'] == 'Online']
top_5_agencies = online_agencies.nlargest(5, 'Count')

fig, ax = plt.subplots()
ax.bar(top_5_agencies['Agency'], top_5_agencies['Count'])
plt.xlabel('Agency')
plt.ylabel('Count')
plt.title('Top 5 Agencies with Highest Online Distribution Channel')

plt.xticks(rotation='vertical')
plt.show()
```

```python
#Since there are only few offline distribution channel, this the pivot table.
dist_agency = merged_df.groupby(['Agency', 'Distribution Channel']).size().reset_index(name='Count')
offline_agencies = dist_agency[dist_agency['Distribution Channel'] == 'Offline']
pivot_data = offline_agencies.pivot(index='Agency', columns='Distribution Channel', values='Count')

pivot_data
```

```python
#PlotCount of Offline Distribution Channel by Agency
dist_agency = merged_df.groupby(['Agency', 'Distribution Channel']).size().reset_index(name='Count')
offline_agencies = dist_agency[dist_agency['Distribution Channel'] == 'Offline']

fig, ax = plt.subplots()
ax.bar(offline_agencies['Agency'], offline_agencies['Count'])
plt.xlabel('Agency')
plt.ylabel('Count')
plt.title('Count of Offline Distribution Channel by Agency')
plt.xticks(rotation='vertical')

plt.show()
```
